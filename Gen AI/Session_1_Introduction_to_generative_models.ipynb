# 2.
Create a table comparing discriminative models and generative models across at least 4 aspects (definition, main use, example task, example model name).

# Comparison of Discriminative Models and Generative Models

| Aspect | Discriminative Models | Generative Models |
|----------|----------------------|------------------|
| **Definition** | Learn the relationship between input data (X) and labels (Y) to predict outcomes. | Learn the underlying data distribution and can generate new data samples similar to the training data. |
| **Main Use** | Classification and prediction tasks. | Data generation, content creation, and modeling data distributions. |
| **Example Task** | Predict whether an email is spam or not spam. | Generate a new email, image, or piece of text similar to existing examples. |
| **Example Model Name** | Logistic Regression, Decision Tree, Support Vector Machine (SVM), ResNet. | GPT, GAN (Generative Adversarial Network), VAE (Variational Autoencoder), DALL·E. |
| **Output** | Predicts a label or class. | Generates new data instances. |
| **Learning Goal** | Learn the decision boundary between classes. | Learn how the data itself is distributed. |

# 4.
Pick one generative AI application (text generation, image synthesis, music/audio generation, or code generation) and describe a new feature you would add to an app you use, explaining how generative AI would power it.

# Generative AI Application: Text Generation

## App Chosen
Spotify

## New Feature: AI Playlist Story Generator

I would add a feature called **AI Playlist Story Generator** to Spotify. After a user creates or listens to a playlist, the feature would generate a short story or description based on the songs, artists, genres, and listening mood.

For example, if a playlist contains relaxing and travel-themed songs, the AI could generate a story such as:

*"This playlist feels like a peaceful road trip through the mountains at sunset, combining soothing melodies and uplifting rhythms to create a calm and adventurous atmosphere."*

## How Generative AI Powers It

This feature would use a **text generation model** (such as GPT) to analyze playlist information, including song titles, genres, artist styles, and listening patterns. The model would then generate unique and personalized descriptions, stories, or mood summaries for each playlist in natural language.

## Benefits

- Makes playlists more engaging and personalized.
- Helps users discover the mood and theme of their playlists.
- Creates a unique listening experience beyond traditional playlist recommendations.
- Encourages users to share playlists with AI-generated descriptions on social media.
