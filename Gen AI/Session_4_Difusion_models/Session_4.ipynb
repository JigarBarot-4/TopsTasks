 # 1.
Simulate the forward diffusion process by writing a Python function that takes a grayscale image (use any small PNG or JPEG) and adds Gaussian noise to it for 10 steps, saving each noisy image to disk.

```python
import cv2
import numpy as np

# Read image in grayscale
img = cv2.imread("imagess/sessipn_4_image.png", 0)

# Convert to float
img = img / 255.0

for i in range(1, 11):
    # Generate Gaussian noise
    noise = np.random.normal(0, 0.05 * i, img.shape)

    # Add noise to image
    noisy_img = img + noise

    # Keep pixel values between 0 and 1
    noisy_img = np.clip(noisy_img, 0, 1)

    # Save image
    cv2.imwrite(f"step_{i}.png", noisy_img * 255)

    print("Saved step", i)

    # Use current noisy image for next step
    img = noisy_img
```

Write a Python script that takes a noisy image and applies a simple denoising filter (like median blur or Gaussian blur) to approximate the reverse diffusion process, then compare the denoised output visually with the original image.<br><br><em><strong>Hint:</strong> Use OpenCV's cv2.medianBlur() or cv2.GaussianBlur() for denoising.</em>

# Denoising a Noisy Image using Gaussian Blur

## Aim
To apply a denoising filter on a noisy image and compare the denoised output with the original image.

## Theory
Noise can degrade the quality of an image. Denoising is the process of removing unwanted noise while preserving important image details. In this experiment, Gaussian Blur is used to reduce noise from a noisy image. This can be considered a simple approximation of the reverse diffusion process, where noise is gradually removed from an image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("imagess/sessipn_4_image.png", 0)

# Add random Gaussian noise
noise = np.random.normal(0, 25, img.shape).astype(np.uint8)
noisy_img = cv2.add(img, noise)

# Apply Gaussian Blur
denoised_img = cv2.GaussianBlur(noisy_img, (5, 5), 0)

# Display images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(noisy_img, cmap='gray')
plt.title("Noisy Image")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(denoised_img, cmap='gray')
plt.title("Denoised Image")
plt.axis('off')

plt.show()
```

## Observation
The noisy image contains random disturbances that affect image quality. After applying Gaussian Blur, the noise is reduced and the image appears smoother.

## Conclusion
Gaussian Blur successfully reduces noise from the image and provides a basic approximation of the reverse diffusion process by removing some of the added noise.

# Risks and Ethical Concerns of Text-to-Image Models

## 1. Creation of Fake or Misleading Images

**Risk:**
Text-to-image models can generate realistic images of events, people, or situations that never actually occurred. Such images may spread misinformation on social media and mislead users.

**Mitigation:**
Developers can add visible or invisible AI-generated watermarks to images so that users can identify them as AI-generated content.

---

## 2. Generation of Harmful or Inappropriate Content

**Risk:**
Users may create offensive, harmful, or inappropriate images that violate community guidelines or negatively impact individuals and groups.

**Mitigation:**
Developers can implement content moderation and safety filters to detect and block harmful prompts and generated images before they are shared.

---

## Conclusion

While text-to-image models such as Stable Diffusion provide powerful creative capabilities, they also introduce ethical challenges. Using watermarking techniques and content moderation systems can help reduce the risks of misinformation and harmful content generation.
