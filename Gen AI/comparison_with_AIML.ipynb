# Comparison of Traditional AI/ML and Generative AI

| Aspect            | Traditional AI/ML                                                                                                                                       | Generative AI (GenAI)                                                                                                                                          |
| ----------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| Data Requirements | Requires labeled and structured data for training and prediction tasks. Example: Spotify uses listening history to recommend songs.                     | Requires very large datasets containing text, images, audio, or videos to generate new content. Example: ChatGPT is trained on large amounts of text data.     |
| Output Type       | Produces predictions, classifications, or recommendations. Example: Instagram recommends posts and reels based on user activity.                        | Creates new content such as text, images, audio, videos, or code. Example: DALL-E generates images from text prompts.                                          |
| Human Interaction | Users usually receive recommendations or automated decisions with limited interaction. Example: Zomato recommends restaurants based on previous orders. | Users interact through prompts and receive customized content. Example: ChatGPT answers questions and generates text based on user input.                      |
| Adaptability      | Designed for specific tasks and performs best within a predefined domain. Example: Spotify's recommendation system focuses on music suggestions.        | Can perform multiple tasks such as writing, summarizing, coding, and image generation. Example: ChatGPT can answer questions, write essays, and generate code. |
| Example Use Cases | Fraud detection, recommendation systems, face recognition, spam filtering. Example: Instagram feed recommendations and Spotify music suggestions.       | Content creation, chatbots, image generation, video summaries, and virtual assistants. Example: ChatGPT, DALL-E, and AI-powered content creation tools.        |
| Creativity        | Does not create original content; mainly analyzes existing data.                                                                                        | Generates original and unique content based on learned patterns.                                                                                               |
| Decision Making   | Focuses on prediction and classification.                                                                                                               | Focuses on content generation and creative problem-solving.                                                                                                    |

## Conclusion

Traditional AI/ML is mainly used for prediction, classification, and recommendation tasks, while Generative AI is designed to create new content such as text, images, and code. Both technologies are widely used in modern applications like Instagram, Spotify, and Zomato to improve user experience and automate tasks.


# 2.
Given a scenario where Zomato recommends restaurants based on your previous orders, explain whether this uses traditional AI/ML or Gen AI, and justify your answer in 3-4 sentences.

# Zomato Restaurant Recommendation Scenario

Zomato recommending restaurants based on a user's previous orders is an example of **Traditional AI/ML** rather than Generative AI. The system analyzes past order history, user preferences, ratings, and browsing behavior to predict which restaurants the user is most likely to choose. It does not create new content such as text or images; instead, it makes recommendations based on patterns found in existing data. Therefore, this use case falls under prediction and recommendation, which are common applications of traditional AI and machine learning.

# Improving Spotify Song Recommendations Using Generative AI

Spotify currently uses traditional AI/ML to recommend songs based on a user's listening history and preferences. With Generative AI, the app could create personalized playlists along with descriptions explaining why each song was selected. It could also generate custom music mixes that match a user's mood, activity, or specific request such as "relaxing songs for studying." In addition, Gen AI could provide natural language interactions, allowing users to ask for music recommendations in a conversational way. This would make the music discovery experience more personalized and interactive.


# 4.
Imagine you are building a chatbot for Flipkart customer support. List two ways a traditional AI/ML approach would handle user queries, and two ways a Gen AI approach would differ in handling the same queries.<br><br><em><strong>Hint:</strong> Focus on how responses are generated and the flexibility of the answers.</em>

# Traditional AI/ML vs Generative AI in a Flipkart Customer Support Chatbot

## Traditional AI/ML Approach

1. The chatbot would identify keywords in the user's query and select a predefined response from a database. For example, if a user asks about order tracking, the chatbot would provide a fixed tracking-related response.

2. The chatbot would work well for common questions such as order status, return policies, or payment issues, but it may struggle to answer questions that are phrased in an unusual way or require detailed explanations.

## Generative AI Approach

1. The chatbot would generate responses dynamically in natural language based on the user's specific question. It could provide personalized and context-aware answers instead of selecting only predefined replies.

2. The chatbot could handle more complex conversations, understand follow-up questions, and explain solutions in different ways. For example, it could guide a user step-by-step through the return process while adapting its response to the conversation.

## Conclusion

Traditional AI/ML chatbots mainly rely on predefined rules and responses, whereas Generative AI chatbots create flexible, human-like responses and can manage more natural and detailed conversations with users.
