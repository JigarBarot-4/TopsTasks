Create a simple Q&A chatbot for Zomato-style restaurant info using any language or platform you know (Python, JavaScript, etc.): given a restaurant name and cuisine, return a canned response like 'Yes, Pizza Palace serves Italian cuisine.'

```python
# Restaurant database
restaurants = {
    "Pizza Palace": "Italian",
    "Spice Garden": "Indian",
    "Dragon Wok": "Chinese",
    "Burger Hub": "American"
}

# Chatbot function
def restaurant_chatbot(restaurant_name, cuisine):
    if restaurant_name in restaurants:
        if restaurants[restaurant_name].lower() == cuisine.lower():
            return f"Yes, {restaurant_name} serves {cuisine} cuisine."
        else:
            return f"No, {restaurant_name} serves {restaurants[restaurant_name]} cuisine, not {cuisine}."
    else:
        return f"Sorry, information for {restaurant_name} is not available."

# Example usage
restaurant = input("Enter restaurant name: ")
cuisine = input("Enter cuisine: ")

response = restaurant_chatbot(restaurant, cuisine)
print(response)
```

Use GitHub Copilot or ChatGPT to help you write a function that takes a list of song names and returns only those that start with the letter 'S', like a Spotify playlist filter.<br><br><em><strong>Constraint:</strong> You must show the prompt you used and the code Copilot or ChatGPT generated.</em>

## prompt

Give me a python function which returns songs which start with s 

```python
def filter_songs_starting_with_s(song_list):
    return [song for song in song_list if song.startswith('S')]

# Example playlist
playlist = [
    "Shape of You",
    "Believer",
    "Senorita",
    "Sunflower",
    "Perfect",
    "Stay"
]

# Filter songs
filtered_playlist = filter_songs_starting_with_s(playlist)

print("Songs starting with 'S':")
print(filtered_playlist)
```

# Production Challenges of Generative AI Chatbots

| Challenge | Description | Possible Solution |
|------------|-------------|------------------|
| **Hallucinations (Incorrect Information)** | The chatbot may generate inaccurate or misleading responses, such as giving wrong movie timings on BookMyShow or incorrect product details on Myntra. | Use **Retrieval-Augmented Generation (RAG)** to fetch information from verified databases and APIs before generating a response. |
| **High Latency and Cost** | Large AI models require significant computing resources, leading to slow response times and high operational costs when serving millions of users. | Use **smaller optimized models**, response caching, and model quantization to reduce inference time and infrastructure costs. |

## Example

- **BookMyShow:** A chatbot might incorrectly state that a movie is available at a theater when it is not. Using real-time theater and ticketing APIs can ensure accurate responses.

- **Myntra:** A chatbot serving millions of customers during a sale may become slow and expensive. Deploying optimized models and caching common queries can improve speed and reduce costs.

## Summary

1. **Hallucinations** → Solve with **RAG and verified data sources**.
2. **High latency/cost** → Solve with **model optimization, caching, and efficient deployment**.
