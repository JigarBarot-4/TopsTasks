Simulate the forward diffusion process by writing a Python function that takes a grayscale image (use any small PNG or JPEG) and adds Gaussian noise to it for 10 steps, saving each noisy image to disk.

```python
import cv2
import numpy as np

# Read image in grayscale
img = cv2.imread("Images/sessipn_4_image.png", 0)

# Convert to float
img = img / 255.0

for i in range(1, 11):
    # Generate Gaussian noise
    noise = np.random.normal(0, 0.05 * i, img.shape)

    # Add noise to image
    noisy_img = img + noise

    # Keep pixel values between 0 and 1
    noisy_img = np.clip(noisy_img, 0, 1)

    # Save image
    cv2.imwrite(f"step_{i}.png", noisy_img * 255)

    print("Saved step", i)

    # Use current noisy image for next step
    img = noisy_img
```
